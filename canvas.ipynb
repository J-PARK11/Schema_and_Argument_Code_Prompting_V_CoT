```python
import os
import json
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Seaborn 스타일 설정
sns.set(style="whitegrid")
colors = sns.color_palette("husl", 40)
```

```python
def plot_pie_chart():
    """
    Train/Valid/Test 퍼즐 데이터를 파이 그래프로 시각화하는 함수
    """
    # 데이터
    labels = ['Train\n(230)', 'Valid\n(20)', 'Test\n(50)']
    sizes = [230, 20, 50]
    colors = ['lightblue', 'lightgreen', 'lightcoral']  # 각 카테고리 색상 지정
    explode = (0.1, 0, 0)  # Train 파이를 약간 띄워서 강조

    # 파이 그래프 그리기
    plt.figure(figsize=(7, 7))
    plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
            shadow=True, startangle=90, textprops={'fontsize': 12})
    
    # 원형 비율 유지
    plt.axis('equal')

    # 제목 추가
    plt.title('17 Train Puzzle Data Split', fontsize=16, fontweight='bold')

    # 그래프 출력
    plt.show()

# 함수 실행
plot_pie_chart()
```

```python
def plot_pie_chart():
    """
    Train/Valid/Test 퍼즐 데이터를 파이 그래프로 시각화하는 함수
    """
    # 데이터
    labels = ['Test\n(300)']
    sizes = [300]
    colors = ['lightyellow']  # 각 카테고리 색상 지정

    # 파이 그래프 그리기
    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, colors=colors,
            shadow=True, startangle=90, textprops={'fontsize': 12})
    
    # 원형 비율 유지
    plt.axis('equal')

    # 제목 추가
    plt.title('3 ZeroShot Test Puzzle', fontsize=16, fontweight='bold')

    # 그래프 출력
    plt.show()

# 함수 실행
plot_pie_chart()
```

```python
loss = os.path.join('/data/jhpark_checkpoint/schema_and_argument_ckpt', 'dump', 'train_loss.json')
with open(loss,'r') as f:
    train_loss = json.load(f)
    print(len(train_loss))
    
loss = os.path.join('/data/jhpark_checkpoint/schema_and_argument_ckpt', 'dump', 'valid_loss.json')
with open(loss,'r') as f:
    valid_loss = json.load(f)
    print(len(valid_loss))
```

```python
def plot_loss(train_loss, val_loss, epochs, len_step_per_epoch, save_path=None):

    # 그래프 그리기
    plt.figure(figsize=(10, 6))
    
    # 학습 로스
    plt.plot(list(map(int,list(train_loss.keys()))), list(train_loss.values()), 'b-', marker='o', label='Training Loss', linewidth=2)
    
    # 검증 로스
    plt.plot(list(map(int,list(valid_loss.keys()))), list(val_loss.values()), 'r-', marker='x', label='Validation Loss', linewidth=2)

    # 그래프 제목과 레이블 설정
    plt.title('Training and Validation Loss', fontsize=14, fontweight='bold')
    plt.xlabel('Steps', fontsize=12)
    plt.ylabel('Loss', fontsize=12)
    
    # 에포크 별 수렴 추이
    # Rectangle((x좌표, y좌표), 가로길이, 세로길이)
    max_value = max(list(val_loss.values()))
    for i in range(epochs):
        rect = patches.Rectangle((len_step_per_epoch*i,0), len_step_per_epoch, max_value, linewidth=0, edgecolor='none', facecolor=colors[i*9], alpha=0.1)
        plt.gca().add_patch(rect)

    # 범례 및 기타 설정
    plt.legend(loc='best', fontsize=12)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()

    # # 저장 경로가 있으면 파일로 저장
    # plt.savefig(save_path, format='png', dpi=300)
    # print(f"Graph saved to {save_path}")
```

```python
plot_loss(train_loss, valid_loss, 4, 120)
```
